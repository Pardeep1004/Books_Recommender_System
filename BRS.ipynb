```python
ratings = pd.read_csv('BX-Book-Ratings.csv', encoding='latin-1', sep = ';', error_bad_lines = False)
ratings
```

```python
ratings.rename(columns={'User-ID': 'user_id','Book-Rating':'rating'},inplace=True)
```

```python
ratings.shape
```

```python
#only 105283 users rate the books out of 278858
ratings['user_id'].value_counts().shape
```

```python
df = ratings['user_id'].value_counts()>200
```

```python
df1 = df[df].index
```

```python
#Now got the user_id
df1
```

```python
#Now only have 526356 ratings(899 users) out of 1149780 ratings(278858 users)
ratings = ratings[ratings['user_id'].isin(df1)]
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on = 'ISBN')
ratings_with_books
```

```python
#Now only have 487671 ratings out of 526356 because we don't have the data of all books
ratings_with_books.shape
# it decreases because some ratings data did not the match or find the books in books dataset
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number_of_rating'},inplace=True)
```

```python
final_rating = ratings_with_books.merge(number_rating, on= 'title')
```

```python
final_rating
```

```python
final_rating.shape
```

```python
print(final_rating.columns)
```

```python
final_rating = final_rating[final_rating['number_of_rating'] >= 50 ]
```

```python
#There are total 61853 ratings of users who did the rating on which books with minimun 50 ratings
final_rating.shape
```

```python
#there are some rows in which users gives ratings to the same books multiple time
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns = 'user_id', index = 'title',values = 'rating')
```

```python
book_pivot
```

```python
#intially we have total 899 users and now left 888 users, the 11 users eliminate because their ratings on books is less than 50
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace= True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
# csr_matrix : It only considers the non zero values to calculate the difference
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm = 'brute')
# algorithm = 'brute' means it calculate the distance from every point
```

```python
model.fit(book_sparse)
```

```python
#np.where(book_pivot.index == 237)
distances = suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1))
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances = suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1))
    
    for i in range (len(suggestions)):
        if i!= 0:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
